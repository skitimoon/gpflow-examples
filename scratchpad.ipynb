```python
import numpy as np
from scipy.stats import wishart, gamma
import matplotlib.pyplot as plt
```

```python
df, scale = 7, 3
```

```python
x = np.linspace(1e-5, 24, 100)
ws = wishart.pdf(x, df=df, scale=scale)  # chi2.pdf(x, df)
print(ws[:5])
plt.plot(x, ws)
```

```python
gm = gamma.pdf(x, a=df/2, scale=scale * 2)
print(gm[:5])
plt.plot(x, gm)
```
